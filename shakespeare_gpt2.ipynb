```
#imports the gpt2-simple python package
import gpt_2_simple as gpt2_simple
from datetime import datetime
```

```
#downloads gpt2 to use and fine tune
gpt2_simple.download_gpt2(model_name='124M')
```

```
#imports google drive to this .ipnyb
from google.colab import files
```

```
#connects this .ipynb to your google drive
gpt2_simple.mount_gdrive()
```

```
#starts the tensorflow training for a gpt2 model
sess = gpt2_simple.start_tf_sess()
```

```
#this finetunes the gpt2 model to the tiny_shakespeare dataset. this takes a while.
#parameters include the number of steps, when to write samples, when to print updates on the loss
gpt2_simple.finetune(sess, dataset="shakespeare.txt", steps=1000, model_name='124M',
                     sample_every=200, save_every=1000, print_every=10, restore_from='fresh')
```

```
#this saves the modeled just trained to google drive
gpt2_simple.copy_checkpoint_to_gdrive(run_name='run1')
```

```
#this generates a text of length x, with a prompt, based on the model just trained.
gpt2_simple.generate(sess, run_name='run1', length=100, prefix="CAESAR:")
```
